## So Now let's Define What's a markov chain
Very Simple : A Markov chain is a stochastic process, but it differs from a general stochastic process in that a Markov chain must be "memory-less". That is, (the probability of) ** future ** actions are not dependent upon the steps that led up to the ** present  ** state.

So the general proprety of a markov chain is  :
$$
\begin{equation*}
P(X_{n+1} = j|X_0 = i_0,X_1 = i_1,....,X_{n-1},X_n =i) =  P(X_{n+1} = j|X_n = i)
\end{equation*}
$$

## Frist thing : Create our Markov Dictionary
** We need to define the dictionary of words **
here i choosed ** ordre = 3 ** which is a parameter that we can tune to get better text generation ,so first we divide our text to words (we call it gram in the fields of computational linguistics and probability ) if a gram is composed by 3 character it's called ** Trigram ** by 2 ** bigram ** and a sequence of n character is called ** n-gram ** ,so we add each character that follows each trigram into an array which is the value in our dictionnary so if a combination of characters (the trigram) is repeated 2 times we will have 2 characters in the array pretty simple ... hein

```python
MarkovList = {}
ordre = 3
for i in range(len(txt) - ordre + 1):
    gram = txt[i:i+ ordre]
    #MarkovList.append(txt[i:i+3])
    if gram not in MarkovList:
        MarkovList[gram] = []
    MarkovList[gram].append(txt[i + ordre : i + ordre + 1])
print(MarkovList)
```

## The generation of text
after difining our states we can build our n-gram model and generate our text ,so first we need to define a starting point ,here we start with the first gram in our dictionary which is ** currentgram ** then with the function ** random ** we pick the next character to add to our result chain (** rslt **) from the array of that gram and we set the range of the for loop to 500 so that the final result will be a 504 character text

```python
import random
def GenrerMarkovText(txt,order,MarkovList):
    currentgram = txt[0:order]
    rslt = currentgram
    for i in range(500):
        possbilite = MarkovList[currentgram]
        nextt = random.choice(possbilite)
        
        #print(rslt)
        rslt +=nextt
        lenth = len(rslt) 
        currentgram = rslt[lenth - ordre:lenth]
        #print('C-'+currentMot)
    print(rslt)
```
